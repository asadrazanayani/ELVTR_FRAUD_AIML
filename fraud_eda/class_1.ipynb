```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = pd.DataFrame({
    'amount': [100, 150, 120, 300, 350, 400, 130, 200, 90, 500],
    'class': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C']
})

# Boxplot
sns.boxplot(x='class', y='amount', data=df)
plt.title('Box Plot of Amount by Class')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)
df = pd.DataFrame({'value': data})

sns.histplot(data=df, x='value', bins=30, kde=True, color='skyblue')
plt.title("Histogram of 'value' with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
df = pd.DataFrame({
    'value': np.concatenate([
        np.random.normal(40, 5, 500),
        np.random.normal(60, 5, 500)
    ]),
    'group': ['A']*500 + ['B']*500
})


```

```python
df.head()
```

```python
sns.histplot(data=df, x='value', hue='group', bins=30, multiple='layer', alpha=0.6)
plt.title('Histogram with Different Colors by Group')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

```python
# Example data
df = pd.DataFrame({
    'math':    [80, 90, 70, 60, 50],
    'science': [88, 85, 60, 65, 55],
    'english': [70, 78, 82, 58, 60],
    'history': [75, 77, 80, 64, 66]
})
```

```python
corr = df.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Sample data
df = sns.load_dataset('tips')

# Violin plot of total_bill by day
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot")
plt.show()
```

```python
sns.kdeplot(data=df, x='total_bill', hue='sex', fill=True)
plt.title("KDE Plot")
plt.show()
```

```python
# Sample DataFrame
data = {
    'date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'attribute_1': [5, 7, 6, 9, 11, 10, 14, 13, 12, 15],
    'attribute_2': [3, 4, 5, 6, 5, 7, 8, 9, 10, 11]
}
df = pd.DataFrame(data)

# Set date as index or use for x-axis
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['attribute_1'], label='Attribute 1')
plt.plot(df['date'], df['attribute_2'], label='Attribute 2')

# Add labels and legend
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Attribute Trends Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
